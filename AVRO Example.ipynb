```python
import avro.schema
import pandas as pd
from avro.datafile import DataFileReader, DataFileWriter
from avro.io import DatumReader, DatumWriter
import collections

schema = avro.schema.Parse(open("user.avsc", "r").read())

User = collections.namedtuple("User", 'name favorite_number favorite_color')
u1 = User(name="Herminio", favorite_number=40, favorite_color="Orange")
writer = DataFileWriter(open("users.avro", "wb"), DatumWriter(), schema)
writer.append({"name": "Alyssa", "favorite_number": 256})
writer.append({"name": "Ben", "favorite_number": 7, "favorite_color": "red"})
writer.append(u1._asdict())
writer.close()

reader = DataFileReader(open("users.avro", "rb"), DatumReader())
userlist = []
for user in reader:
    userlist.append(user)
reader.close()
```

```python
userlist
```

```python
df = pd.DataFrame(userlist)
```
